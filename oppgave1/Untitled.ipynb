## Finding the Polar Angle $\theta$ from Cartesian Coordinates
Given a complex number $\text {z = x + iy}$, the goal is to determine its angle $\theta$ in polar form. 
The angle is computed using the inverse tangent function and adjusted based on the quadrant.

### Case 1: When \( x > 0 \)
- **If \( y > 0 \) (First Quadrant)**: $$\theta = \arctan\left(\frac{y}{x}\right)$$
- **If \( y < 0 \) (Fourth Quadrant)**: $$\theta = \arctan\left(\frac{y}{x}\right) + 2\pi$$
  Since $\arctan(y/x)$ returns a negative value, adding $2\pi$ ensures the angle is in the range $ [0, 2\pi] $.
- **If \( y = 0 \) (Positive Real Axis)**: $$\theta = 0$$
### Case 2: When \( x < 0 \)
- **For all \( y \) values (Second and Third Quadrants)**: $$\theta = \arctan\left(\frac{y}{x}\right) + \pi$$
  Since $\arctan(y/x)$ alone would return an angle in the incorrect quadrant, adding $\pi$ shifts it accordingly.

### Case 3: When \( x = 0 \)
- **If \( y > 0 \) (Positive Imaginary Axis)**:
  $$\theta = \frac{\pi}{2}$$
- **If \( y < 0 \) (Negative Imaginary Axis)**:
  $$\theta = -\frac{\pi}{2}$$
- **If \( y = 0 \) (Origin, Undefined Angle)**:
  $$\theta \text{ is undefined}$$

```python
import numpy as np 
import math as m



def tetha_add( real, imaginary):
        
        while(real > 0):
            if(imaginary > 0):
                tetha = m.atan(imaginary/real)
                break
            elif(imaginary < 0):
                tetha = m.atan(imaginary/real) + 2*m.pi
                break 
            elif(imaginary == 0):
                tetha = 0
                break
            else:
                break
        while(real < 0):
            tetha = m.atan(imaginary/ real) + m.pi 
            break
        while(real == 0):
            if(imaginary < 0):
                tetha =  -m.pi/2
                break
            elif(imaginary > 0):
                tetha = m.pi/2
                break
            elif(imaginary == 0):
                tetha = None
                break
        return tetha
```

## Ploting the function 
### Here we user the Angle in Radian with the help of tetha_add(real, imaginary) the atan2() function of python do the same. atan without (2) is the invers tan

```python
import math
import matplotlib.pyplot as plt

def plot_complex_solutions(n, solutions):
    """
    Plot the solutions on the complex plane and label each point with its angle.
    Also draw lines from the origin to the points.
    
    :param n: An integer (e.g., the exponent) - used for the plot title.
    :param solutions: A list of tuples (x, y), each representing a point in the complex plane.
    """
    plt.figure(figsize=(6, 6))

 
    for (x, y) in solutions:
        plt.plot([0, x], [0, y], color="blue", linestyle="-", linewidth=1)
        plt.scatter(x, y, color="red")
        angle_radians = tetha_add(y, x)  
        angle_degrees = math.degrees(angle_radians)
        text_offset = 0.02
        plt.text(x + text_offset, y + text_offset, 
                 f"{angle_degrees:.2f}°", fontsize=9)

  
    plt.axhline(0, color='black', linewidth=0.8)
    plt.axvline(0, color='black', linewidth=0.8)
    
   
    plt.gca().set_aspect('equal', adjustable='box')

   
    plt.title(f"{n} Solutions on the Complex Plane")
    plt.grid(True)
    plt.show()
```
